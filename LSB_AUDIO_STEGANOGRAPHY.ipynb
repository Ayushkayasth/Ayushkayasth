```python
import wave
```

```python
def read_audio(filename):
  with wave.open(filename, mode='rb') as wav_file: #'rb' is using for read but in binary format
    #audio parameter
    num_channels = wav_file.getnchannels()
    sample_width = wav_file.getsampwidth()
    frame_rate = wav_file.getframerate()
    frames = wav_file.getnframes()

    data = wav_file.readframes(frames)
  return data, num_channels, sample_width, frame_rate
```

```python
cover_data, num_channels, sample_width, frame_rate = read_audio("/content/drive/MyDrive/Ayush_Cyber/CantinaBand3.wav")
```

```python
secret_text = input("Enter the Message:")
binary_secret = ''.join(format(ord(char), '08b') for char in secret_text)
print(binary_secret)
```

```python
bin_sec_list = list(binary_secret)
print(bin_sec_list)
print("No. of bits in above binary code : ",len(bin_sec_list))
```

```python
required_size = len(binary_secret)
if required_size > len(cover_data):
   print("Audio file is not large enough to hold the secret text.")
```

```python
cover_data_list = list(cover_data)
for i in range(required_size):
    cover_data_list[i] = (cover_data_list[i] & 0xFE) | int(binary_secret[i])
cover_data = bytes(cover_data_list)
print(cover_data)
```

```python
def bytes_to_binary(byte_data):
  # Convert each byte to its binary representation (8 bits)
  binary_string = ''.join(format(b, '08b') for b in byte_data)
  return binary_string

bin_cover = bytes_to_binary(cover_data)
print("Converted binary string:", bin_cover)
```

```python
def write_audio(filename, data, num_channels, sample_width, frame_rate):
  with wave.open(filename, mode='wb') as wav_file:
    wav_file.setnchannels(num_channels)
    wav_file.setsampwidth(sample_width)
    wav_file.setframerate(frame_rate)
    wav_file.writeframes(data)
    wav_file.writeframesraw(data)

write_audio("stego_audio.wav", cover_data, num_channels, sample_width, frame_rate)
```

```python
def encode_text(audio_file):
      with open(audio_file, 'rb') as audio:
          audio_data = audio.read()

      # Extract secret text from audio data using LSB
      binary_secret = ''

      for byte in audio_data:
          binary_secret += str(byte & 0x01)
      return binary_secret

bs = encode_text("/content/output_data_file.wav")
print(bs)
```

```python
def compare_binary_strings(str1, str2):
  # Ensure both strings have the same length
  # Count the number of differing bits
  num_diff = 0
  bin_list = []
  for i in range(len(str1)):
    if str1[i] == str2[i]:
      num_diff += 1
      bin_list.append(str1[i])

  return num_diff,bin_list
num_same,bin_list = compare_binary_strings(binary_secret, bs)

print("Number of Same bits:", num_same)
print(bin_list)
```

```python
#program to convert a list to string

def listToString(s):
	str1 = ""
	for ele in s:
		str1 += ele
	return str1

final_string = listToString(bin_list)
print(final_string)
```

```python
def binary_to_text(binary_string):
  # Check if the string length is a multiple of 8 (8 bits per character)
  if len(binary_string) % 8 != 0:
    print("Error: Binary string length must be a multiple of 8.")
    return None

  # Convert the binary string to groups of 8 bits (bytes)
  byte_list = [binary_string[i:i+8] for i in range(0, len(binary_string), 8)]

  # Convert each byte to its corresponding character (assuming ASCII)
  text = ''.join(chr(int(byte, 2)) for byte in byte_list)

  return text

# Example usage
text = binary_to_text(final_string)

print("Hidden Message is:", text)
```
